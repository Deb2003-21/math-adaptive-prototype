```python
#real time tracker 

import matplotlib.pyplot as plt

'''correct = [1, 1, 0, 1, 0, 1, 1, 0, 1, 1]
time = [2, 3, 5, 4, 6, 3, 2, 7, 4, 3]
'''
def track(correct,time1):
    # Pie Chart
    correct_count = sum(correct)
    incorrect_count = len(correct) - correct_count

    plt.figure()
    plt.pie(
        [correct_count, incorrect_count],
        labels=["Correct", "Incorrect"],
        autopct="%1.1f%%"
    )
    plt.title("Accuracy Summary (Total Questions: 10)")
    plt.show()

    # Line Chart
    plt.figure()
    plt.plot(range(1, 11), time1, marker='o')
    plt.xlabel("Question Number")
    plt.ylabel("Time Taken (seconds)")
    plt.title("Time Taken per Question")
    plt.show()
```
