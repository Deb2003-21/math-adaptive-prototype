```python

def difficulty_modifier(ans,t,total):
    #  maintaing history
    
    count_ones = lambda arr: sum(arr)
    n=count_ones(ans)
    # maintaining difficulty
    if ((n//total)*100)==100 and int(t)<=5:
        level=5
    elif ((n//total)*100)==80 and int(t)>=7:
        level=4
    elif int(t)>=9 and int(t)<=12 and ((n//total)*100)>=30:
        level=3
    elif int(t)>=14:
        level=2
    else: 
        level=1

    #returning history if total question=10
    return level

    
```
